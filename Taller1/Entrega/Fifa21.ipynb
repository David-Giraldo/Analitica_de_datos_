##Regresión Lineal

Es una técnica paramétrica de machine learning, que consiste en encontrar cuáles son los mejores parámetros (coeficientes) para los datos que tengamos. Los mejores coeficientes serán los que minimicen alguna medida de error, además, en este caso se usa el error cuadrático medio.

\begin{equation}
y=wx+b
\end{equation}

Para la regresión lineal se usa **sklearn.linear_model.LinearRegression**, el cual es un algoritmo de aprendizaje automático basado en el aprendizaje supervisado, que modela un valor de predicción basado en variables independientes, de esta manera  determina la relación entre las variables y la previsión.


El código para realizar una regresión lineal es:

***class sklearn.linear_model.LinearRegression( * , fit_intercept = True , normalize = False , copy_X = True , n_jobs = None , positive = False )***

##Regresión Ridge

La Regresión Rigde, también denominada regresión contraída o Tikhonov regularization, regulariza el modelo resultante imponiendo una penalización al tamaño de los coeficientes de la relación lineal entre las características predictivas y la variable objetivo. En este caso, los coeficientes calculados minimizan la suma de los cuadrados de los residuos penalizada al añadir el cuadrado de la norma L2 del vector formado por los coeficientes:

\begin{equation}
RSS_{ridge}=\sum_{i=1}^{n}(y_{i}-f(x_{i}))^2+\lambda\sum_{j=1}^{p}\beta_{j}^{2}
\end{equation}

donde $\lambda$ es un parámetro que controla el grado de penalización: cuanto mayor sea, los coeficientes serán menores resultando más robustos a la colinealidad.

Para la regresión de Ridge se usa **sklearn.linear_model.Ridge**,
es un tipo de regresión lineal utilizado para resolver un modelo de regresión donde la función de pérdida es la función de mínimos cuadrados lineales y la regularización es $L_2$. Este, reduce los coeficientes para las variables de entrada que no contribuyen mucho en la predicción.

El código para realizar una regresión de Ridge es:

***class sklearn.linear_model.Ridge(alpha=1.0, *, fit_intercept=True, normalize=False, copy_X=True, max_iter=None, tol=0.001, solver='auto', random_state=None)***

##Regresión Lasso

Lasso (Least Absolute Shrinkage and Selection Operator) es un modelo lineal que penaliza el vector de coeficientes añadiendo su norma L1 (basada en la distancia Manhattan) a la función de coste:

\begin{equation}
RSS_{lasso}=\sum_{i=1}^{n}(y_{i}-f(x_{i}))^2+\lambda\sum_{j=1}^{p}|\beta_{j}|
\end{equation}

Lasso tiende a generar "coeficientes dispersos": vectores de coeficientes en los que la mayoría de ellos toman el valor cero, esto quiere decir que el modelo va a ignorar algunas de las características predictivas, lo que puede ser considerado un tipo de selección automática de características, de esta manera, al incluir menos características supone un modelo más sencillo de interpretar que puede poner de manifiesto las características más importantes del conjunto de datos. 

Para la regresión de Lasso se usa **sklearn.linear_model.Lasso**
es una modificación de la regresión lineal,la función de pérdida se modifica para minimizar la complejidad del modelo limitando la suma de los valores absolutos de los coeficientes del modelo, además se hace el uso de una restricción de la norma $L_1$ con el fin de reducir los coeficientes para aquellas variables de entrada que no contribuyen mucho a la tarea de predicción.

\begin{equation}
(1 / (2 * n_{samples})) * ||y - Xw||^2_2 + alpha * ||w||_1$
\end{equation}

El código para realizar una regresión lineal de Lasso es:

**class sklearn.linear_model.Lasso( alpha = 1.0 , * , fit_intercept = True , normalize = False , precompute = False , copy_X = True , max_iter = 1000 , tol = 0.0001 , warm_start = False , positivo = False , random_state = Ninguno , selección = 'cíclico' )**

##Regresión ElasticNet

Es un modelo de regresión lineal que normaliza el vector de coeficientes con las normas L1 y L2. Esto permite generar un modelo en el que solo algunos de los coeficientes sean no nulos, manteniendo las propiedades de regularización de Ridge. La función de coste es equivalente a:

\begin{equation}
RSS_{Enet}=\sum_{i=1}^{n}(y_{i}-f(x_{i}))^2+\alpha(\lambda\sum_{j=1}^{p}\beta_{j}^2+(1-\lambda)\sum_{j=1}^{p}|\beta_{j}|)
\end{equation}

El parámetro $\lambda$ regula el peso dado a la regularización impuesta por Ridge y por Lasso. Desde este punto de vista Elastic Net es un superconjunto de ambos modelos.

Para la regresión ElasticNet se usa **sklearn.linear_model.ElasticNet**, ese logaritmo es útil cuando hay varias funciones correlacionadas pues elije todas las características, tal combinación permite aprender un modelo disperso donde pocos de los pesos son distintos de cero como el método de regularización de Lasso.

\begin{equation}
1 / (2 * n_samples) * ||y - Xw||^2_2
+ alpha * l1_ratio * ||w||_1
+ 0.5 * alpha * (1 - l_{1_{ratio}}) * ||w||^2_2
\end{equation}

El código para realizar una regresión lineal ElasticNet es:

**class sklearn.linear_model.ElasticNet(alpha=1.0, *, l1_ratio=0.5, fit_intercept=True, normalize=False, precompute=False, max_iter=1000, copy_X=True, tol=0.0001, warm_start=False, positive=False, random_state=None, selection='cyclic')**

##Regresión KernelRidge

La regresión de la cresta de Kernel (KRR) combina la regresión Ridge y la clasificación por mínimos cuadrados lineales con regularización de norma  L2  con el kernel trick. De este modo, aprende una función lineal en el espacio inducida por el núcleo respectivo y los datos. Para los núcleos no lineales, esto corresponde a una función no lineal en el espacio original.

Para la regresión KernelRidge se usa **sklearn.kernel_ridge.KernelRidge**, la cual trata de una regresión no lineal, el cual es más eficiente cuando se trabaja con datos complejos, puesto a que las regresiones lineales a menudo no son suficientes para captar algunas peculiaridades del problema, a lo que la no linealidad podría ser favorable.
La regresión aprende una función lineal en el espacio inducido por el kernel respectivo y los datos, sin embargo,Kernels no lineales, va a corresponder a funciones no lineales en el espacio original.

El código para realizar una regresión KernelRidge es:

**class sklearn.kernel_ridge.KernelRidge(alpha=1, *, kernel='linear', gamma=None, degree=3, coef0=1, kernel_params=None)**

```python
#%%   
#codificar estadísticas -> jugadores fifa ej 88+3
def code_stats(y):
    yc = np.zeros(y.shape[0])
    for i in range(y.shape[0]):
        #print(y.iloc[i])
        
        if y.iloc[i].find("+") > -1: # encontrar signo mas en str y casteo a flotante
            yc[i] = float(y.iloc[i][:y.iloc[i].find("+")])+float(y.iloc[i][y.iloc[i].find("+")+1:])
        else: yc[i] = float(y.iloc[i])
        #print(yc[i])
    return yc   

#codificar moneda ->
def code_euro(y):
    yc = np.zeros(y.shape[0])
    for i in range(y.shape[0]):
      #  print(i,y[i])
        if y[i][-1]=='M': yc[i] = float(y[i][1:-1])*10**6 #buscar M y reemplazar 10^6
        elif y[i][-1]=='K': yc[i] = float(y[i][1:-1])*10**3 # buscar K y reemplazar por 10^3
        else: yc[i] = float(y[i][1:])
        #print(yc[i])
    return yc 

#codificar hits ->
def code_hits(y):
    yc = np.zeros(y.shape[0])
    for i in range(y.shape[0]):
      #  print(i,y[i])
        y[i]=str(y[i])
        if y[i][-1]=='M': yc[i] = float(y[i][0:-1])*10**6 #buscar M y reemplazar 10^6
        elif y[i][-1]=='K': yc[i] = float(y[i][0:-1])*10**3 # buscar K y reemplazar por 10^3
        else: yc[i] = float(y[i][0:])
        #print(yc[i])
    return yc 
```

```python
#%% dummy transformer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder

class dummy_fifa(BaseEstimator,TransformerMixin):
    #inicializacion de clase y varaibles
    def __init__(self, col_euro,col_stats, cat_usr): #constructor clase
        self.col_euro = col_euro #lista atributos tipo moneda
        self.col_stats = col_stats #lista atributos tipo estadistica
        self.cat_usr = cat_usr #lista de atributos categoricos

    def fit(self,X, *_):
        Xi = X.copy() #copiar dataset para no reemplazar original
        self.imputer_num = SimpleImputer(strategy="most_frequent") #crear imputador tipo modo
        self.a = Xi.columns[np.sum(Xi.isna())> 0] #encontrar columnas con datos faltantes
        #print(a)
        self.imputer_num.fit(Xi[self.a]) # ajustar imputador
        
        Xi[self.a] = self.imputer_num.transform(Xi[self.a]) #evaluar  sobre datos imputador
        #print('Cod Euros\n')
        for i in self.col_euro: #codificar tipo moneda
         #   print(i)
            Xi[i] = code_euro(np.array(Xi[i]))
        #print('Cod stats\n')    
        for i in self.col_stats: #codificar datos estadisticos
         #   print(i)
            Xi[i] = code_stats(Xi[i])
        
        for i in ['Hits']: 
         #   print(i)
            Xi[i] = code_hits(np.array(Xi[i]))
            
        
        #height, wieght corregir formato
        Xi['Height'].replace(regex=["'"], value='.',inplace=True) 
        Xi['Height'].replace(regex=['"'], value='',inplace=True)        
        for i in Xi.index:
            #print(float(Xi.loc[i,'Weight'][:-3]))
            Xi.loc[i,'Weight'] = float(Xi.loc[i,'Weight'][:-3])
            Xi.loc[i,'Height'] = float(Xi.loc[i,'Height'])
        
        Xi['Height'] = Xi['Height'].astype('float64');#print(Xi['Height'].dtype)

        Xi['Joined'] = Xi['Joined'].replace(regex="Jan",value="01")
        Xi['Joined'] = Xi['Joined'].replace(regex="Feb",value="02")
        Xi['Joined'] = Xi['Joined'].replace(regex="Mar",value="03")
        Xi['Joined'] = Xi['Joined'].replace(regex="Apr",value="04")
        Xi['Joined'] = Xi['Joined'].replace(regex="May",value="05")
        Xi['Joined'] = Xi['Joined'].replace(regex="Jun",value="06")
        Xi['Joined'] = Xi['Joined'].replace(regex="Jul",value="07")
        Xi['Joined'] = Xi['Joined'].replace(regex="Aug",value="08")
        Xi['Joined'] = Xi['Joined'].replace(regex="Sep",value="09")
        Xi['Joined'] = Xi['Joined'].replace(regex="Oct",value="10")
        Xi['Joined'] = Xi['Joined'].replace(regex="Nov",value="11")
        Xi['Joined'] = Xi['Joined'].replace(regex="Dec",value="12")    
        Xi['Joined'] = Xi['Joined'].replace(regex=",",value=".")
        Xi['Joined'] = Xi['Joined'].replace(regex=" ",value="")
        Xi['Joined'] = Xi['Joined'].astype('float64')

        Xi['W/F'] = Xi['W/F'].replace(regex=" ★",value="")
        Xi['W/F'] = Xi['W/F'].replace(regex="★",value="")
        Xi['W/F'] = Xi['W/F'].astype('float64')

        Xi['SM'] = Xi['SM'].replace(regex=" ★",value="")
        Xi['SM'] = Xi['SM'].replace(regex="★",value="")
        Xi['SM'] = Xi['SM'].astype('float64')

        Xi['IR'] = Xi['IR'].replace(regex=" ★",value="")
        Xi['IR'] = Xi['IR'].replace(regex="★",value="")
        Xi['IR'] = Xi['IR'].astype('float64')
        

        cat = [] #codificar variables categoricas con ordinal encoder
        for i in self.cat_usr.keys():
            cat = cat + [[*self.cat_usr.get(i)]]
        self.col_cat_usr = OrdinalEncoder(categories=cat)
        Xi[[*self.cat_usr.keys()]] =self.col_cat_usr.fit_transform(Xi[[*self.cat_usr.keys()]])
        
        #self.col_cat = Xi.columns[Xi.dtypes=='O']
        #self.cod = OrdinalEncoder()
        #self.cod.fit(Xi[self.col_cat])
        
        return self    

    def transform(self, X, *_):#funcion transformador
        Xi = X.copy()
        Xi[self.a] = self.imputer_num.transform(Xi[self.a])
        
        for i in self.col_euro:
         #   print(i)
            Xi[i] = code_euro(np.array(Xi[i]))
        #print('Cod stats\n')    
        for i in self.col_stats:
         #   print(i)
            Xi[i] = code_stats(Xi[i])

        for i in ['Hits']: 
         #   print(i)
            Xi[i] = code_hits(np.array(Xi[i]))
        
        #height, wieght
        Xi['Height'].replace(regex=["'"], value='.',inplace=True)
        Xi['Height'].replace(regex=['"'], value='',inplace=True)        
        for i in Xi.index:
            #print(float(Xi.loc[i,'Weight'][:-3]))
            Xi.loc[i,'Weight'] = float(Xi.loc[i,'Weight'][:-3])
            Xi.loc[i,'Height'] = float(Xi.loc[i,'Height'])
        
        Xi['Height'] = Xi['Height'].astype('float64')
        Xi['Weight'] = Xi['Weight'].astype('float64')

        Xi['Joined'] = Xi['Joined'].replace(regex="Jan",value="01")
        Xi['Joined'] = Xi['Joined'].replace(regex="Feb",value="02")
        Xi['Joined'] = Xi['Joined'].replace(regex="Mar",value="03")
        Xi['Joined'] = Xi['Joined'].replace(regex="Apr",value="04")
        Xi['Joined'] = Xi['Joined'].replace(regex="May",value="05")
        Xi['Joined'] = Xi['Joined'].replace(regex="Jun",value="06")
        Xi['Joined'] = Xi['Joined'].replace(regex="Jul",value="07")
        Xi['Joined'] = Xi['Joined'].replace(regex="Aug",value="08")
        Xi['Joined'] = Xi['Joined'].replace(regex="Sep",value="09")
        Xi['Joined'] = Xi['Joined'].replace(regex="Oct",value="10")
        Xi['Joined'] = Xi['Joined'].replace(regex="Nov",value="11")
        Xi['Joined'] = Xi['Joined'].replace(regex="Dec",value="12")    
        Xi['Joined'] = Xi['Joined'].replace(regex=",",value=".")
        Xi['Joined'] = Xi['Joined'].replace(regex=" ",value="")
        Xi['Joined'] = Xi['Joined'].astype('float64')

        Xi['W/F'] = Xi['W/F'].replace(regex=" ★",value="")
        Xi['W/F'] = Xi['W/F'].replace(regex="★",value="")
        Xi['W/F'] = Xi['W/F'].astype('float64')

        Xi['SM'] = Xi['SM'].replace(regex=" ★",value="")
        Xi['SM'] = Xi['SM'].replace(regex="★",value="")
        Xi['SM'] = Xi['SM'].astype('float64')

        Xi['IR'] = Xi['IR'].replace(regex=" ★",value="")
        Xi['IR'] = Xi['IR'].replace(regex="★",value="")
        Xi['IR'] = Xi['IR'].astype('float64')
        
        Xi[[*self.cat_usr.keys()]] =self.col_cat_usr.transform(Xi[[*self.cat_usr.keys()]])
        #Xi[self.col_cat]= self.cod.transform(Xi[self.col_cat])
        return Xi
    
    def fit_transform(self,X,*_):
        self.fit(X)
        return self.transform(X)
```
